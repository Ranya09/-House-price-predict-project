```python
import pandas as pd

# Remplacez les zéros par NaN dans la colonne 'sqft_basement'
df['price'].replace(0, pd.NA, inplace=True)

# Calculez la médiane de la colonne 'sqft_basement' en excluant les NaN
median_basement = df['price'].median()

# Remplacez les NaN par la médiane calculée
df['price'].fillna(median_basement, inplace=True)


print(df.head())
```

```python
import pandas as pd

# Remplacez les zéros par NaN dans la colonne 'sqft_basement'
df['yr_renovated'].replace(0, pd.NA, inplace=True)

# Calculez la médiane de la colonne 'sqft_basement' en excluant les NaN
median_basement = df['yr_renovated'].median()

# Remplacez les NaN par la médiane calculée
df['yr_renovated'].fillna(median_basement, inplace=True)


print(df.head())
```

```python
import pandas as pd

# Supposons que vous ayez déjà chargé vos données dans un DataFrame df
# Si ce n'est pas le cas, assurez-vous de charger vos données en conséquence.

# Filtrer les valeurs positives dans la colonne 'Quality'
positive_quality = df[df['Quality'] > 0]

# Compter les occurrences des valeurs positives de 'Quality'
quality_occurrences = positive_quality['Quality'].value_counts()

# Grouper le DataFrame par 'Quality' et calculer la moyenne du 'Price' pour chaque groupe
price_mean_by_quality = positive_quality.groupby('Quality')['price'].mean()

# Afficher le nombre d'occurrences des valeurs positives de 'Quality'
print(f"Occurrences des valeurs positives dans la colonne 'Quality':\n{quality_occurrences}\n")

# Afficher la moyenne du 'Price' pour chaque valeur positive de 'Quality'
print(f"Moyenne du 'Price' par 'Quality':\n{price_mean_by_quality}\n")
```

```python

# Filtrer les données pour inclure uniquement les maisons avec 'quality' strictement supérieur à 0
filtered_data = df[df['Quality'] > 0]
filtered_data = filtered_data[filtered_data['waterfront'] == 1]
filtered_data = filtered_data[filtered_data['condition'] > 2 ]



# Sélectionner les colonnes 'bedrooms', 'bathrooms', 'price' et 'id' pour le DataFrame filtré
selected_columns = ['bedrooms', 'bathrooms', 'price', 'id', 'sqft_above', 'sqft_basement' ,'sqft_lot', 'condition', 'view', 'Quality', 'city',  'floors', 'waterfront' ]
filtered_data = filtered_data[selected_columns]

# Trier le DataFrame par ordre croissant de prix
filtered_data = filtered_data.sort_values(by='price')

# Afficher le tableau résultant
print(filtered_data)
```

```python
# Grouper les données par 'bedrooms' et 'bathrooms' et calculer le prix minimum
min_price_by_combination = filtered_data.groupby(['bedrooms', 'bathrooms'])['price'].min().reset_index()

# Calculer la qualité maximale pour chaque combinaison de 'bedrooms' et 'bathrooms'
max_quality_by_combination = filtered_data.groupby(['bedrooms', 'bathrooms'])['Quality'].max().reset_index()

# Afficher le DataFrame résultant
print(min_price_by_combination)
print(max_quality_by_combination)
```
